```python
# В названиях столбцов pcp 01, pcp 06, pcp 24 встречается пробел.
# Заменен пробел в названиях на знак нижнего подчёркивания. 

taxi = taxi.rename(columns={'pcp 01': 'pcp_01', 'pcp 06': 'pcp_06', 'pcp 24': 'pcp_24'})
taxi.head()
```

```python
# Из какого района (borough) было совершено наибольшее количество поездок (pickups) за весь период.

taxi.groupby('borough', as_index = False).agg({'pickups' : 'sum'})\
    .rename(columns = {'pickups' : 'sum_pickups'})\
    .sort_values(by = 'sum_pickups', ascending=False)
```

Наибольшее количество поездок совершено с района Manhattan

```python
# название района с наименьшим числом поездок 

min_pickups = taxi.groupby('borough').agg({'pickups' : 'sum'}).idxmin()
min_pickups 
```

Наименьшее количество поездок совершено с района EWR

```python
 # Отобрать данные о поездках в выходные дни ('Y')
    
query1 = taxi.query("hday == 'Y'")
query1.head()
```

```python
# Сравнить среднее число поездок, 
# и выбрать районы, из которых по праздникам ('Y') в среднем поступает больше заказов, чем в обычные дни.
```

```python
new = taxi.groupby(['borough', 'hday']).agg({'pickups' : 'mean'}).sort_values( by = 'pickups', ascending =False)
new
```

В праздничные дни больше всего заказов поступает из районов Queens и EWR

```python
# Для каждого района посчитайте число поездок по месяцам. Определить месяц и район с максимальным количеством поездок.

pickups_by_mon_bor = taxi.groupby(['pickup_month', 'borough'])\
                     .agg({'pickups' : 'sum'})\
                     .sort_values('pickups', ascending = False)\
                     .reset_index()
pickups_by_mon_bor
```

Максимальное число поездок было в июне с района Manhattan

```python
# Функция, которая получает на вход колонку с температурой в °F и возвращает значения, переведённые в градусы Цельсия.

def temp_to_celsius(temp_farenheit):
    return (temp_farenheit - 32) * 5 / 9
С = temp_to_celsius(taxi.temp)
С
```
